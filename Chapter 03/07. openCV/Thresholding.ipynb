```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('fry.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

retval, dst = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
plt.imshow(dst, cmap='gray')
```

```python
img = cv.imread('dog.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

ret, otsu = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

plt.imshow(otsu, cmap = 'gray')
```

```python
img2 = cv.imread('notes.png')
img2_gray = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

dst = cv.adaptiveThreshold(img2_gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 9, 3)

plt.figure(figsize=(20, 15))
plt.imshow(dst, cmap='gray')
```

```python
kernel = cv.getStructuringElement(cv.MORPH_CROSS, (3,3))

ret, threshold = cv.threshold(img, 200, 255, cv.THRESH_BINARY)

dilation = cv.dilate(threshold, kernel, iterations=4)
plt.imshow(dilation, cmap='gray')
```
