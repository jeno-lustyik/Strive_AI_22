```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('m.jpg')
hsv_img = cv.imread('hsv.png')
```

```python
hsv_img = cv.cvtColor(hsv_img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(20, 15))
plt.imshow(hsv_img)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
lower_blue = np.array([85, 50, 50])
upper_blue = np.array([130, 255, 255])

mask_blue = cv.inRange(img_hsv, lower_blue, upper_blue)

res_blue = cv.bitwise_and(img_rgb, img_rgb, mask=mask_blue)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(res_blue)
```

```python
lower_green = np.array([40, 50, 50])
upper_green = np.array([70, 255, 255])

mask_green = cv.inRange(img_hsv, lower_green, upper_green)

res_green = cv.bitwise_and(img_rgb, img_rgb, mask=mask_green)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(res_green)
```

```python
# lower_gb = np.array([40, 50, 50])
# upper_gb = np.array([130, 255, 255])

mask_gb = cv.bitwise_or(mask_blue, mask_green)

res_gb = cv.bitwise_and(img_rgb, img_rgb, mask=mask_gb)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(mask_gb)
plt.imshow(res_gb)
```

```python
lower_red1 = np.array([0,50,50])
upper_red1 = np.array([10,255,255])
lower_red2 = np.array([170,50,50])
upper_red2 = np.array([180,255,255])

mask_red1 = cv.inRange(img_hsv, lower_red1, upper_red1)
mask_red2 = cv.inRange(img_hsv, lower_red2, upper_red2)

mask_red = mask_red1 + mask_red2
mask_rgb = cv.bitwise_or(mask_red, mask_gb)
res_rgb = cv.bitwise_and(img_rgb, img_rgb, mask=mask_rgb)

plt.figure(figsize=(10, 6))
plt.imshow(res_rgb)
```

```python
res_final = cv.addWeighted(res_rgb, 0.8, img_rgb, 0.2, 0)
plt.figure(figsize=(10, 6))
plt.imshow(res_final)
```

```python
img2 = cv.imread('mm_large.png')
img2_rgb = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
img2_hsv = cv.cvtColor(img2, cv.COLOR_BGR2HSV)

lower_gb = np.array([40, 50, 50])
upper_gb = np.array([130, 255, 255])

lower_red1 = np.array([0,50,50])
upper_red1 = np.array([10,255,255])
lower_red2 = np.array([170,50,50])
upper_red2 = np.array([180,255,255])

mask_red1 = cv.inRange(img2_hsv, lower_red1, upper_red1)
mask_red2 = cv.inRange(img2_hsv, lower_red2, upper_red2)

mask_gb = cv.inRange(img2_hsv, lower_gb, upper_gb)
mask_red = mask_red1 + mask_red2

mask_rgb = cv.bitwise_or(mask_gb, mask_red)
res_gb = cv.bitwise_and(img2_rgb, img2_rgb, mask=mask_rgb)
plt.figure(figsize=(10, 6))
plt.imshow(res_gb)
```
