```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('fry.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
copy_img = img.copy()

pts = np.array([[10,300],[300,200],[70,20]])

cv.polylines(copy_img, [pts], True, (255, 0, 0), 3)

plt.imshow(copy_img)
```

```python
copy_img = img.copy()

pts = np.array([[200,350],[400,350],[400,100],[200,100]])
pts.reshape((-1, 1, 2))

cv.polylines(copy_img, [pts], True, (255, 0, 0), 3)

plt.imshow(copy_img)
```
